# Scraping do site da Honda

## Importação de bibliotecas

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
url="https://www.honda.com.br/motos/modelos"
response=urlopen(url)
html=response.read()
soup=BeautifulSoup(html, "html.parser")

```

## Listas de valores e caracteristicas de cada moto

```python
#Listas de cada caracteristicas de cada moto
valores_nomes=[]
valores_precos=[]
cilindradas=[]
cor=[]
nacionalidade=[]
categoria_=[]
#Dicionario para adicionar essas informaçoes para fazer um dataframe
motos={}
```

## Scraping dos dados das motos

```python
soup.find_all("form", {"action": "/"})
for c in soup.find_all("form", {"action": "/"}):
    print(c)
```

```python
#adicionando os valores dos nomes de cada moto na lista de nomes
for c in soup.find_all("form", {"action": "/"}):
    for name in c.find_all("input", {"name":"name"}):
        valores_nomes.append(name.get("value"))
```

```python
#adicionando os valores de cada categoria de cada moto na listas de categorias
for c in soup.find_all("form", {"action": "/"}):
    for categoria in c.find_all("input", {"name":"categoria"}):
        categoria_.append(categoria.get('value'))
```

```python
#adicionando os precos de cada moto  na listas de preços
for preco in soup.find_all("span",{"class":"price"}):
    if preco.get_text() !="R$ 50.300":#foi feito essa condição pois essa valor estava sem as outras caracteristicas da moto, logo decidi não adicionar a lista
        valores_precos.append(preco.get_text())

```

```python
#adcicionando os valores das cilindradas de cada moto na lista de cilinnndradas
for c in soup.find_all("form", {"action": "/"}):
    for cilindro in c.find_all("input", {"name": "cilindradas"}):
        cilindradas.append(cilindro.get("value"))
```

```python
#adicionando a nacionalidade de cada moto na lista de nascionalidades
for c in soup.find_all("form", {"action": "/"}):
    for butao in (c.find_all("button", {"class":"btn highlight-button modal-interesse-button"})):
        nacionalidade.append(butao.get("data-categoria"))
```

```python
#adicionando as cores de cada moto em uma lista de cores
for c in soup.find_all("form", {"action": "/"}):
    for butao in (c.find_all("button", {"data-click":"tenhointeresse"})):
        cor.append(butao.get("data-color"))

```

## Criando o dicionário com os valores coletados do site da honda

```python
#adicionando cada lista ao dicionario
motos["Nome"]=valores_nomes
motos["nacionalidade"]=nacionalidade
motos["categoria"]=categoria_
motos["cor"]=cor
motos["cilindradas"]=cilindradas
motos["preços"]=valores_precos
motos
```

## Criando o dataframe

```python
df=pd.DataFrame(motos)
```

## Transformando o dataframe em um arquivo CSV para fazer uma analise e tratamento dos dados coletados

```python
df.to_csv("Motos_honda.csv", index=False)
```
